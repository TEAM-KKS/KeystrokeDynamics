```python
import pandas as pd

key_stroke_df = pd.read_csv("keyStrokeData.csv")
```

```python
key_stroke_df.columns
```

```python
key_stroke_df.head()
```

```python
key_stroke_df['false_character'] = (key_stroke_df['false_character'] == 't').astype(int)
key_stroke_df.head()
```

```python
key_stroke_df['name'] = (key_stroke_df['name'] == 'Kedar').astype(int)
key_stroke_df.head()
# Kedar -> 1 and Sope -> 0
```

```python
key_storke_df_kedar = key_stroke_df.loc[key_stroke_df['name'] == 1]
key_storke_df_kedar.describe()
```

```python
key_storke_df_Sope = key_stroke_df.loc[key_stroke_df['name'] == 0]
key_storke_df_Sope.describe()
```

```python
test_df = key_stroke_df.loc[key_stroke_df['box_id'] == '#text-area-test']
train_df = key_stroke_df.loc[key_stroke_df['box_id'] != '#text-area-test']
```

```python
test_df.describe()
```

```python
train_df.describe()
```

```python
train_df_x = train_df[['false_character', 'hold_for', 'location', 'long_pressed_equivalent', 'pressed_after']]
train_df_y = train_df[['name']]
test_df_x = test_df[['false_character', 'hold_for', 'location', 'long_pressed_equivalent', 'pressed_after']]
test_df_y = test_df[['name']]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt

%matplotlib inline

logreg = LogisticRegression()
logreg.fit(train_df_x, train_df_y)
```

```python
pred_df_y = logreg.predict(test_df_x)
```

```python
cnf_matrix = metrics.confusion_matrix(test_df_y, pred_df_y)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(test_df_y, pred_df_y))
print("Precision:",metrics.precision_score(test_df_y, pred_df_y))
print("Recall:",metrics.recall_score(test_df_y, pred_df_y))
```

```python
from sklearn.model_selection import train_test_split 

X = key_stroke_df[['false_character', 'hold_for', 'location', 'long_pressed_equivalent', 'pressed_after']]
y = key_stroke_df[['name']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler

scale_df = key_stroke_df[['false_character', 'hold_for', 'location', 'long_pressed_equivalent', 'pressed_after']]
x = scale_df.values
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA().fit(x)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance')
```

```python
pca = PCA(n_components = 5)
pca_components = pca.fit_transform(x)
pca_df = pd.DataFrame(data=pca_components, columns=["PC1", "PC2", "PC3", "PC4", "PC5"])
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca=train_test_split(pca_df,y,test_size=0.20,random_state=0)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_pca, y_train_pca)
y_pred_pca = logreg.predict(X_test_pca)
cnf_matrix = metrics.confusion_matrix(y_test_pca, y_pred_pca)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test_pca, y_pred_pca))
print("Precision:",metrics.precision_score(y_test_pca, y_pred_pca))
print("Recall:",metrics.recall_score(y_test_pca, y_pred_pca))
```
